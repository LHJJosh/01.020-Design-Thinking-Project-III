```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

url ='https://drive.google.com/file/d/1i8Q53ynwSD3bI3arzr2W-ekPGeieSUfy/view?usp=sharing'
url2='https://drive.google.com/uc?id=' + url.split('/')[-2]
df =  pd.read_csv(url2)
display(df)

def normalize_z(df):
    mean = df.mean(axis=0)
    std = df.std(axis=0)
    dfout = (df-mean)/std
    return dfout
    pass

def get_features_targets(df, feature_names, target_names):
    df_feature = df.loc[:,feature_names]
    df_target = df.loc[:,target_names]
    return df_feature, df_target

def prepare_feature(df_feature):
    feature = df_feature.to_numpy()
    array1 = np.ones((feature.shape[0],1))
    X = np.concatenate((array1,feature), axis = 1)
    return X

def prepare_target(df_target):
    return df_target.to_numpy()
    pass

def predict(df_feature, beta):
    feature_norm = normalize_z(df_feature)
    X = prepare_feature(feature_norm)
    return predict_norm(X,beta)

def predict_norm(X, beta):
    return np.matmul(X,beta)

def r2_score(y, ypred):
    #calculate ssres
    diff = y - ypred
    ssres = np.matmul(diff.T, diff)[0][0]
    #calculate sstot
    ymean = np.mean(y)
    diff_mean = y - ymean
    sstot = np.matmul(diff_mean.T, diff_mean)[0][0]
    #calculate r2
    return 1 - (ssres/sstot)
    pass

def mean_squared_error(target, pred):
    m = target.shape[0]
    error = target - pred
    return (1/m) * np.sum(error**2)
    pass

def compute_cost(X, y, beta):
    J = 0
    #calculate m, num of rows/data points
    m = y.shape[0]
    #calculate the yp, predicted target value from X and beta
    yp = np.matmul(X,beta)
    #calculate the error
    error = yp -y
    #calculate the cost
    J = (1/(2*m))* np.matmul(error.T, error)
    J = J[0][0]
    return J

def gradient_descent(X, y, beta, alpha, num_iters):
    #calculate m from the shape of X or y
    m = X.shape[0]
    J_storage = np.zeros(num_iters)
    #for the number of iterations
    for n in range(num_iters):
        #->compute the predicted y
        yp = np.matmul(X, beta)
        #->compute the error
        error = yp -y
        #->compute the new beta
        beta = beta - (alpha/m) * np.matmul(X.T, error)
        #->compute J using the new beta and store it
        J_storage[n] =  compute_cost(X, y, beta)
    return beta, J_storage
```

```python
#List of all continents
ls = ["Africa", "Asia", "Europe", "North America", "Oceania", "South America"]
#For new cases variable
#i is continent; for continent in ls
for i in ls:
    df_2 = df.loc[(df['continent'] == i), :]    
    df_features, df_target = get_features_targets(df_2, ["new_cases", "avg_stringency_idx", "Days"], ["new deaths"])
    # Normalize using z normalization
    df_features = normalize_z(df_features)
    X = prepare_feature(df_features)
    target = prepare_target(df_target)
    #Changed from 1000 to 20000 due to accuracy problems
    iterations = 1500
    alpha = 0.01
    beta = np.zeros((4,1))

    beta, J_storage = gradient_descent(X, target, beta, alpha, iterations)
    plt.plot(J_storage)

    # Call predict()
    pred = predict(df_features, beta)

    # Change target to numpy array
    target = prepare_target(df_target)

    # Calculate r2 score by calling a function
    r2 = r2_score(target, pred)
    txt = "For {i}, the coefficient for new cases, average stringency index, days and the intercept are respectively: {total_deaths_coefficient}, {avg_stringency_idx_coefficient}, {days_coefficient}, {intercept}."
    r2_text = "The r^2 value for {i} is {r_2}."
    print(txt.format(i = i, total_deaths_coefficient = beta[1], avg_stringency_idx_coefficient = beta[2], days_coefficient = beta[3], intercept = beta[0]) )
    print(r2_text.format(i = i, r_2 = r2))
    print()
```

```python
# Functions used to access and prepare features and target data

# Function that carries out z-normalization on a Dataframe
def normalize_z(df):
    mean = df.mean(axis=0)
    std = df.std(axis=0)
    dfout = (df-mean)/std
    return dfout
    pass

# Function that obtains the Feature and Target Dataframes from a Dataframe
def get_features_targets(df, feature_names, target_names):
    df_feature = df.loc[:,feature_names]
    df_target = df.loc[:,target_names]
    return df_feature, df_target

# Function that prepares and converts Feature Dataframe into a Numpy array
def prepare_feature(df_feature):
    feature = df_feature.to_numpy()
    array1 = np.ones((feature.shape[0],1))
    X = np.concatenate((array1,feature), axis = 1)
    return X

# Function that converts the Target Dataframe into a Numpy array
def prepare_target(df_target):
    return df_target.to_numpy()
    pass

# Function used to split dataset into training and test datasets randomly
def split_data(df_feature, df_target, random_state=None, test_size=0.5):

    feature_indices = df_feature.index
    target_indices = df_target.index

    if random_state != None:
        np.random.seed(random_state)

    #k is the number of rows in test set
    num_rows = len(feature_indices)
    k = int(test_size * num_rows)
    test_indices = np.random.choice(feature_indices, k, replace = False)
    indices = set(feature_indices)
    test_indices = set(test_indices)
    train_indices = indices - test_indices
    df_feature_train = df_feature.loc[train_indices,:]
    df_target_train = df_target.loc[train_indices,:]
    df_feature_test = df_feature.loc[test_indices,:]
    df_target_test = df_target.loc[test_indices,:]

    return df_feature_train, df_feature_test, df_target_train, df_target_test

# Obtaining Feature and Target Dataframes:
features = ["new_cases","1 - vacc_rate"]
target = ["icu_usage"]
task2_df_feature, task2_df_target = get_features_targets(task2_df, features, target)

# Splitting dataset into Train and Test datasets:
df_feature_train, df_feature_test, df_target_train, df_target_test = split_data(task2_df_feature,task2_df_target,random_state=100, test_size=0.3)

# Normalizing training features:
df_feature_train_z = normalize_z(df_feature_train)

# Preparing and changing training features into np array:
X = prepare_feature(df_feature_train_z)

# Changing target into np array:
target = prepare_target(df_target_train)
```

```python
# Functions used to build multiple linear regression model

# Function that calculates the Cost Function of a linear regression model
def compute_cost(X, y, beta):
    J = 0
    #calculate m, num of rows/data points
    m = y.shape[0]
    #calculate the yp, predicted target value from X and beta
    yp = np.matmul(X,beta)
    #calculate the error
    error = yp -y
    #calculate the cost
    J = (1/(2*m))* np.matmul(error.T, error)
    J = J[0][0]
    return J

# Function that iterates the beta values in order to minimize the Cost Function of a Model
def gradient_descent(X, y, beta, alpha, num_iters):
    #calculate m from the shape of X or y
    m = X.shape[0]
    J_storage = np.zeros(num_iters)
    #for the number of iterations
    for n in range(num_iters):
        #->compute the predicted y
        yp = np.matmul(X, beta)
        #->compute the error
        error = yp -y
        #->compute the new beta
        beta = beta - (alpha/m) * np.matmul(X.T, error)
        #->compute J using the new beta and store it
        J_storage[n] =  compute_cost(X, y, beta)
    return beta, J_storage

# Initial beta values set to zeroes:
beta = np.zeros((X.shape[1],1))

# Configuring alpha and number of iterations to build our model over:
alpha = 0.01
iterations = 1000

# Obtaining final beta values and cost our built model:
beta, J_storage = gradient_descent(X, target, beta, alpha, iterations)
print(f"beta values = {beta}")

plt.plot(J_storage)
```

```python
# Function that predicts for the dependent variable given the predictor variables:
def predict(df_feature, beta):
    feature_norm = normalize_z(df_feature)
    X = prepare_feature(feature_norm)
    return np.matmul(X,beta)

# Function that calculates the r2 score for a built model:
def r2_score(y, ypred):
    #calculate ssres
    diff = y - ypred
    ssres = np.matmul(diff.T, diff)[0][0]
    #calculate sstot
    ymean = np.mean(y)
    diff_mean = y - ymean
    sstot = np.matmul(diff_mean.T, diff_mean)[0][0]
    #calculate r2
    return 1 - (ssres/sstot)

# Function that calcualtes the mean squared error for a built model:
def mean_squared_error(target, pred):
    m = target.shape[0]
    error = target - pred
    return (1/m) * np.sum(error**2)

# Function that calculates the adjusted r2 score for a built model:
def adjusted_r2_score(target, pred, num_of_features):
    m = target.shape[0]
    r2 = r2_score(target, pred)
    adjust_r2 = 1 - (1-r2) * (m-1) / (m-num_of_features-1)
    return adjust_r2

# Function that calculates the root mean squared error for a built model:
def root_mean_squared_error(mse):
  return np.sqrt(mse)

# Calling predict():
pred_test= predict(df_feature_test, beta)
pred_train = predict(df_feature_train,beta)

# Changing target to numpy array:
target_test = prepare_target(df_target_test)
target_train = prepare_target(df_target_train)

# Calculating r2 score of built model:
r2_test = r2_score(target_test, pred_test)
r2_train = r2_score(target_train, pred_train)

# Calculating mse of built model:
mse_test = mean_squared_error(target_test,pred_test)
mse_train = mean_squared_error(target_train, pred_train)

# Calculating Adjusted r2 score of built model:
adjust_r2_test = adjusted_r2_score(target_test, pred_test, len(features))
adjust_r2_train = adjusted_r2_score(target_train, pred_train, len(features))

# Calculating rmse of built model:
rmse_test = root_mean_squared_error(mse_test)
rmse_train = root_mean_squared_error(mse_train)

print(f"r2 for test: {r2_test}, r2 for train: {r2_train}")
print(f"Adjusted r2 for test: {adjust_r2_test}, Adjusted r2 for train: {adjust_r2_train}")
print(f"mse for test: {mse_test}, mse for train: {mse_train}")
print(f"rmse for test: {rmse_test}, rmse for train: {rmse_train}")
```

We came up with 6 metrics to evaluate our model's accuracy:
- The **r2 scores** measure the percentage of the variance in the dependent variable that the independent variables explain collectively, and tells us about the extent and direction of linear correlation between the dependent and independent variables. However, the r2 score is not the best metric to measure linear correlation for a multiple linear regression model since the more predictor variables used to build the model, the higher the r2 score will be. Hence, the adjusted r2 score will be a better metric.
- The **adjusted r2 scores** are a modified version of r2 scores that have been adjusted for the number of predictors in the model. The adjusted R-squared increases only if the new term improves the model more than would be expected by chance. It decreases when a predictor improves the model by less than expected by chance. Hence, the adjusted r2 scores are a good metric used to compare multiple linear regression models with different numbers of predictor variables.
- The **mean squared errors (mse)** measure how close a fitted line is to data points by calculating the mean of the squared vertical distances between each actual data point and the corresponding point on our model's curve. Hence the mse is a good indicator of how accurate our model is in predicting the ICU usage rate of Singapore given the predictor variables of our model.
- The **root mean squared errors (rmse)** is the square root of the mse, and tell us the average distance of an actual data point from the fitted line, measured along a vertical line, and is directly interpretable in terms of measurement units. Hence the rmse will be a better metric compared to the mse.

We measured each of these metrics to our training and test datasets. Since the model was built on the training dataset, the metrics are likely to be higher on the training dataset compared to the testing dataset. Hence, we are more concerned with the metrics used on the testing dataset.

Therefore, we will be focusing and commenting on the **adjusted r2 scores on our test sets** and the **rmse values on our test sets** of each our models.

The Improved Model 4 is already the model with the best performance. However the performance of model 4 on the test set is apparently worse than on training set,i.e. mse for Improved Model 4 on training set = 0.0021929961560782614, mse for Improved Model 4 on test set = 0.007629258469254766
This may be caused by the high variance problem(overfitting). So we can apply L2 normalization based on model 4


```python
def compute_cost_L2(X, y, beta, lambdavalue):
    J = 0
    #calculate m, num of rows/data points
    m = y.shape[0]
    #calculate the yp, predicted target value from X and beta
    yp = np.matmul(X,beta)
    #calculate the error
    error = yp -y
    #calculate the L2 term
    L2_reg_term = (lambdavalue / 2 * m) * np.sum(np.square(beta))
    #calculate the cost
    J = (1/(2*m))* np.matmul(error.T, error) + L2_reg_term
    J = J[0][0]
    return J

def gradient_descent_L2(X, y, beta, alpha, num_iters, lambdavalue):
    #calculate m from the shape of X or y
    m = X.shape[0]
    J_storage = np.zeros(num_iters)
    #for the number of iterations
    for n in range(num_iters):
        #->compute the predicted y
        yp = np.matmul(X, beta)
        #->compute the error
        error = yp -y
        #->compute the new beta
        beta = (1 - alpha * lambdavalue / m) * beta - (alpha / m) * np.matmul(X.T, error)
        #->compute J using the new beta and store it
        J_storage[n] =  compute_cost_L2(X, y, beta, lambdavalue)
    return beta, J_storage
```
